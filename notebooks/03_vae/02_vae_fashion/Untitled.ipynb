```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(666)
x = np.random.rand(100, 1)*2
y=3*x +4+(np.random.rand(100,1)-0.5)
y_=3*x+4
plt.scatter(x,y,marker='x',color='red',alpha=0.5)
plt.plot(x,y_,color='blue',)
```

```python
x.shape
```

```python
import torch
```

```python
from torch.utils.data import  DataLoader ,TensorDataset
```

```python
x=torch.from_numpy(x).float()
y=torch.from_numpy(y).float()
```

```python
dataset=TensorDataset(x,y)
dataloader=DataLoader(dataset,batch_size=16,shuffle=True)
print("len of dataloader(数据集被分为多少组):%d"%len(dataloader))
for index,(data,label) in enumerate(dataloader):
    print("index= %d size_per_batch= %d "%(index,len(data)))
```

```python
print(dir(dataset))
```

```python
w=torch.randn(1,requires_grad=True)
b=torch.randn(1,requires_grad=True)
for epoch in range(1,51):
    for batch_i , (data,label) in enumerate(dataloader):
        h=x*w+b
        print(data.shape)
        loss=torch.mean((h-y)**2)
        print("%.3lf"%loss.item())
        loss.backward()
        w.data-=0.01*w.grad.data
        b.data-=0.01*b.grad.data
        w.grad.zero_()
        b.grad.zero_()
        print("epoch(%d) batch(%d) loss=%.3lf"%(epoch,batch_i,loss.item()))
```
